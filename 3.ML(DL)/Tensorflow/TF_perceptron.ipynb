# 1. 퍼셉트론

```python
import tensorflow as tf
```

```python
tf.random.set_seed(777) # 시드 설정
```

```python
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses import mse
```

```python
# 데이터 준비하기
x = np.array([[0,0],[1,0],[0,1],[1,1]])
y = np.array([[0],[1],[1],[1]])
```

```python
# 모델 구성하기
model = Sequential()
```

```python
# 단층 퍼셉트론 구성
model.add(Dense(1, input_shape = (2,), activation = 'linear'))
# 1 = 퍼셉트론 갯수(은닉유닛), (2, ) = 두개의 특성을 가진 1차원의 데이터
```

```python
model.compile(optimizer = SGD(),
             loss = mse,
             metrics = ['acc']) # list 형태로 평가지표 표시
```

```python
model.fit(x,y, epochs = 500)
```

```python
model.get_weights()
```

```python
# 2개는 가중치, 1개는 편향(bias) : 올바른 개수의 가중치를 얻음 (OR 게이트 문제 해결)
```

## 활성화함수

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# 시그모이드 함수
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 하이퍼볼릭 탄젠트 함수
def tanh(x):
    return list(map(lambda x: math.tanh(x), x))

# relu 함수
def relu(x):
    result = []
    for ele in x:
        if(ele <= 0):
            result.append(0)
        else:
            result.append(ele)
            
    return result
```

```python
# 시그모이드 함수 그려보기
x = np.linspace(-4, 4, 100)
sig = sigmoid(x)
plt.plot(x, sig)
plt.show()
```

```python
tanh = tanh(x)
plt.plot(x, tanh)
plt.show()
```

```python
relu = relu(x)
plt.plot(x, relu)
plt.show()
```

# 경사하강법

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lr_list = [0.001, 0.1, 0.5, 0.9]
```

```python
def get_derivative(lr):
    w_old = 2
    derivative = [w_old]
    # 손실함수
    y = [w_old ** 2]
    
    for i in range(1, 10):
        # 미분값 구하기
        dev_value = w_old * 2
        # 가중치 업데이트
        w_new = w_old - lr * dev_value
        w_old = w_new
        
        derivative.append(w_old)
        y.append(w_old ** 2)
        
    return derivative, y
```

```python
x = np.linspace(-2, 2, 50)
x_square = [i ** 2 for i in x]

fig = plt.figure(figsize = (12, 7))

for i, lr in enumerate(lr_list):
    derivative, y = get_derivative(lr)
    ax = fig.add_subplot(2, 2, i+1)
    ax.scatter(derivative, y, color = 'red')
    ax.plot(x, x_square)
    ax.title.set_text('lr = ' + str(lr))

plt.show()
```
